## A Review of the .apply ( ) Method on Single Columns

```python
df = pd.read_csv("jamesbond.csv", index_col='Film')
df.sort_index(inplace=True)
df.head(3)
```

```python
def cnv_to_str_and_add_millions(number):
    return str(number) + " MILLIONS!"

df["Box Office"] = df["Box Office"].apply(cnv_to_str_and_add_millions)
```

```python
df.Budget = df.Budget.apply(cnv_to_str_and_add_millions)
```

```python
df.head(3)
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    df[col] = df[col].apply(cnv_to_str_and_add_millions)
```

```python
df.sample(3)
```
